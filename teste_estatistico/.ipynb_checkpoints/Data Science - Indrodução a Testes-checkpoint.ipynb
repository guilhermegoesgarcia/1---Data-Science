## Aplicando os  conceitos de confiança

### Vamos utilizar os conceitos de confiabilidade para definir se a média das nossas amostras são confiaveis para definir as médias de amostras maiores do que a nossa..  Definindo por exeplo se a nossa média de notas pode representar a média de notas do mundo todo

```python
from statsmodels.stats.weightstats import zconfint
```

```python
zconfint(nota_media_dos_filmes_com_pelomenos_10_votos)
```

### Imaginando que temos uma amostra grande o suficiente para representar os nossos dados e imaginando que os filmes em $Geral$ tenham $Caracteristicas$ similares ao do estudado 
### Então: temos um intervalo de confiança de nota entre ( 3.41  ---  3.45 )

#### Obs: existe outros tipos de testes ..  ( normalmente o T teste é usado patra amostragem pequena )

```python
from statsmodels.stats.weightstats import DescrStatsW
```

```python
descr_todos_com_10_votos = DescrStatsW(nota_media_dos_filmes_com_pelomenos_10_votos)
descr_todos_com_10_votos.tconfint_mean()
```

```python
round(notas1.rating.mean(), 3) # média do filme toy history
```

```python
zconfint(notas1.rating) # calculo de confiabilidade do filme
```

```python
from statsmodels.stats.weightstats import ztest
ztest(notas1.rating, value = 3.4320503405352603 )
```

#### como o valor é menor que 0.05 podemos concluir que as médias não são iguais

### Observando a o calculo de confiabilidade o gráfico e o calculo de pvalue .. concluimos que sim, os dados são diferentes .. ja que todos os valores são diferentes do calculado anteriormente

```python
np.random.seed(75243)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i] , value = 3.4320503405352603)
    return (i,media,p)

valores= np.array([calcula_teste(i)  for i in range(2, len(temp))])  # uma outra forma de escrever o comando For.




fig = plt.figure(figsize=(10,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(valores[:,0],valores[:,1])
eixo.plot(valores[:,0],valores[:,2])
eixo.hlines(y=0.05,xmin=2,xmax=len(temp),color='k', linestyle='--')
eixo.legend(['Média','Pvelue'])
```

#### Obs: z-teste e t-teste devem ser usado em distribuições normais 

### Caso a distribuição não seja normal é necessario normaliza-la

# -Teste para definir se os dados se comportam como normal-

```python
from scipy.stats import normaltest

stat, p = normaltest(notas1.rating)
p
```

#### menor que 0,05 informa que os dados não são normais

```python
# esse teste é não parametrico

from scipy.stats import ranksums

_, p = ranksums(notas1.rating, notas593.rating)
p
```

#### Descartamos a hipotese nula.. hipótese nula é de que as duas amostras vieram da mesma distribuição. Já a hipótese alternativa, que admitiremos nesse caso (já que o p value foi bem pequeno), é a de que os valores de uma amostra têm uma tendência a serem maiores que os da outra amostra

#### Ou seja, se pegarmos notas aleatórias para Toy Story e Silence of the Lambs, as notas deste último, em geral, são maiores.
