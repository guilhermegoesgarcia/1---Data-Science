# Aula 3 - pontos de máximo e minimo ( add seta)

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df.data, df.temperatura, color='g')
eixo.set_title('Temperatuda do Momento', fontsize=25, pad=20)
eixo.set_xlabel('Data',fontsize=20)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.legend(['Temperatura'],loc='lower right',fontsize=15)
eixo.grid(True)


eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')  # apresentação de eixos máximo
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--') # apresentação de eixos minimo
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df.data, df.temperatura, color='g')
eixo.set_title('Temperatuda do Momento', fontsize=25, pad=20)
eixo.set_xlabel('Data',fontsize=20)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.legend(['Temperatura'],loc='lower right',fontsize=15)
eixo.grid(True)

x1=df.data[df.temperatura.idxmax()] # determinando o ponto máximo
y1=max(df['temperatura'])

eixo.annotate('Máximo',xy=(x1,y1),fontsize=20) # escrevendo msg no ponto máximo

eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')  # apresentação de eixos máximo
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--') # apresentação de eixos minimo
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df.data, df.temperatura, color='g')
eixo.set_title('Temperatuda do Momento', fontsize=25, pad=20)
eixo.set_xlabel('Data',fontsize=20)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.legend(['Temperatura'],loc='lower right',fontsize=15)
eixo.grid(True)

x1=df.data[df.temperatura.idxmax()] # determinando o ponto máximo
y1=max(df['temperatura'])

x2 = df.data[df.temperatura.idxmax() - 7000] # local de inicio da seta 
y2 = max(df['temperatura']) - 5

eixo.annotate('Máximo',xy=(x1,y1),fontsize=20,
             xytext =(x2,y2),arrowprops=dict(facecolor='k')) # escrevendo msg no ponto de inicio da seta

eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')  #
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--') # 
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df.data, df.temperatura, color='g')
eixo.set_title('Temperatuda do Momento', fontsize=25, pad=20)
eixo.set_xlabel('Data',fontsize=20)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.legend(['Temperatura'],loc='lower right',fontsize=15)
eixo.grid(True)


x1=df.data[df.temperatura.idxmax()] # determinando o ponto máximo
y1=max(df['temperatura'])

x2 = df.data[df.temperatura.idxmax() - 7000] # local de inicio da seta 
y2 = max(df['temperatura']) - 5

eixo.annotate('Máximo',xy=(x1,y1),fontsize=20,
             xytext =(x2,y2),arrowprops=dict(facecolor='k')) # escrevendo msg no ponto de inicio da seta

eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')  #
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--') # 




x1=df.data[df.temperatura.idxmin()] # determinando o ponto Minimo
y1=min(df['temperatura'])

x2 = df.data[df.temperatura.idxmin() - 7000] # local de inicio da seta 
y2 = min(df['temperatura']) + 5

eixo.annotate('Mínimo',xy=(x1,y1),fontsize=20,
             xytext =(x2,y2),arrowprops=dict(facecolor='k')) # escrevendo msg no ponto de inicio da seta
```

```python
temperatura_dia_da_semana = temperatura_dia_da_semana[nome_dias]
temperatura_dia_da_semana
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores={'Iris-setosa':'r','Iris-versicolor':'b','Iris-virginica':'g'}
for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'],tmp['largura_sépala'],
                 color=cores[especie])


eixo.set_title('Gráfico de Dispersão',fontsize=25,pad=15)
eixo.set_xlabel('Comprimento da Sépala',fontsize=15)
eixo.set_ylabel('Largura da Sépala',fontsize=15)
eixo.tick_params(labelsize=15)
eixo.legend(cores,fontsize=20)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores={'Iris-setosa':'r','Iris-versicolor':'b','Iris-virginica':'g'}
marcadores = {'Iris-setosa':'x','Iris-versicolor':'o','Iris-virginica':'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'],tmp['largura_sépala'],
                 color=cores[especie],marker=marcadores[especie])


eixo.set_title('Gráfico de Dispersão',fontsize=25,pad=15)
eixo.set_xlabel('Comprimento da Sépala',fontsize=15)
eixo.set_ylabel('Largura da Sépala',fontsize=15)
eixo.tick_params(labelsize=15)
eixo.legend(cores,fontsize=20)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores={'Iris-setosa':'r','Iris-versicolor':'b','Iris-virginica':'g'}
marcadores = {'Iris-setosa':'x','Iris-versicolor':'o','Iris-virginica':'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'],tmp['largura_sépala'],
                 color=cores[especie],marker=marcadores[especie],
                s=100)


eixo.set_title('Gráfico de Dispersão',fontsize=25,pad=15)
eixo.set_xlabel('Comprimento da Sépala',fontsize=15)
eixo.set_ylabel('Largura da Sépala',fontsize=15)
eixo.tick_params(labelsize=15)
eixo.legend(cores,fontsize=20)
```

# Aula 5 - Add Anotações

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu,sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'],bins=20)
eixo.set_title('Histograma', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala',fontsize=15)
eixo.grid(True)

eixo.annotate('$mu= {0:0.2f}$\n$ sigma={1:.2f}$'.format(mu,sigma),xy=(4.5, 20), fontsize=20) # anotação
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu,sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'],bins=20)
eixo.set_title('Histograma', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala',fontsize=15)
eixo.grid(True)

eixo.annotate('$mu= {0:0.2f}$\n$ sigma={1:.2f}$'.format(mu,sigma),xy=(4.5, 20), fontsize=20) # anotação

eixo.axvline(mu, color='k',linestyle='--') # linha vertical para identificar a média
eixo.annotate('média',xy=(mu-1.3,28),fontsize=20)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu,sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'],bins=20)
eixo.set_title('Histograma', fontsize=15)
eixo.set_xlabel('Comprimento da Pétala',fontsize=15)
eixo.grid(True)

eixo.annotate('$\mu= {0:0.2f}$\n$ \sigma={1:.2f}$'.format(mu,sigma),xy=(4.5, 20), fontsize=20) # anotação

eixo.axvline(mu, color='k',linestyle='--') # linha vertical para identificar a média
eixo.annotate('média',xy=(mu-1.3,28),fontsize=20) # descritivo ao lado

eixo.axvline(df['comprimento_pétala'].median(), color='g',linestyle='--') # linha vertical para identificar a mediana
eixo.annotate('mediana',xy=(df['comprimento_pétala'].median()+0.1,28),fontsize=20) # descritivo ao lado
```

### Combinando as 3 imagens geradas anteriormente

```python
from PIL import Image

largura,altura = Image.open('Histograma_ivc.png').size
combinada = Image.new("RGB",(3*largura,altura))
intervalo = 0

for imagem in map(Image.open,['Histograma_iv.png','Histograma_is.png','Histograma_ivc.png']):
    combinada.paste(imagem,(intervalo, 0))
    intervalo += largura
    
combinada.save('combinada.png')
```
